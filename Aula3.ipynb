```python
#Exercício 01 - while
usuario = input("Digite o nome do usuário: ")
senha = input ("Digite a senha: ")

while senha == usuario:
  print("Atenção: senha não pode ser igual ao nome do usuário. Digite novamente!")
  usuario = input("Digite o nome do usuário: ")
  senha = input("Digite a senha: ")
print("Cadastro realizado com sucesso! =)")
```

```python
#Exercício 02 - while
n = int(input("Digite um número: "))
soma = 0
numero = 1
#n = 5
while numero <=n:
  if numero % 2 == 0:
    soma +=numero #soma = soma + numero
  numero +=1

print(f"A soma dos números pares entre 1 e {n} é {soma}")
```

```python
#Exercício 05 - while
#Requisitos do Programa:
#O programa deve solicitar ao usuário que insira a duração de cada treino em minutos.
#O usuário pode encerrar o registro do dia digitando "0".
#O programa deve calcular e exibir o total de minutos treinados.
#O programa deve calcular e exibir a média de duração dos treinos.

total_minutos = 0
quantidade_treinos = 0

duracao = int(input("Digite a duração do seu treino em minutos: "))
while duracao!=0:
  total_minutos+=duracao #total_minutos = total_minutos + duracao
  quantidade_treinos +=1
  duracao = int(input("Digite a duração do seu treino em minutos: "))

if quantidade_treinos > 0:
  media = total_minutos/quantidade_treinos
  print("##### Resultado do seu Treino ####")
  print(f"O total de minutos treinados {total_minutos}")
  print(f"Média de duração dos seus treinos {round(media,2)} em minutos") #round para arredondar valores
else:
  print("Nenhum treino registrado! Bora treinar meu povo =)")
```

```python
#Exercício 02 - for

#Um petshop atende 5 cachorros por tarde. Faça um programa que solicite ao usuário o código do serviço efetuado:
#1 - banho;
#2 - tosa;
#3 - banho e tosa;
#4- outros).

#Por fim, exiba a quantidade de solicitações para cada um dos serviços.

banho = 0
tosa = 0
banho_tosa = 0
outros = 0
print("###### Serviços do petshop #####")
print(" 1- Banho, 2 - Tosa , 3- Banho e Tosa e 4 - Outros")
for i in range(1,6):
  codigo = int(input(f"Digite o código do serviço do {i}º cachorro: "))
  if codigo == 1:
    banho +=1
  elif codigo == 2:
    tosa +=1
  elif codigo == 3:
    banho_tosa +=1
  else:
    outros +=1

print("###### Resumo das Solicitações ######")
print(f"Banho {banho}")
print(f"Tosa {tosa}")
print(f"Banho e Tosa {banho_tosa}")
print(f"Outros {outros}")
```
